```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Using sklearn's standard dataset load_breast_cancer
```

```python
from sklearn.datasets import load_breast_cancer
```

```python

cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['target'])
```

```python
print(cancer['feature_names'])
```

```python
cancer['data'].shape
```

```python
df_c = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns= np.append(cancer['feature_names'],['target']))
```

```python
df_c.head(30)
```

```python
#DATA VISUALIZATION:
```

```python
sns.countplot(df_c['target'])
```

```python
sns.scatterplot(x='mean area',y='mean smoothness',hue='target',data =df_c)
```

```python
#SPLITTING DATASET:
```

```python
x = df_c.drop(['target'],axis =1)
```

```python
y= df_c['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2 ,random_state=10)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
#Using DecisionTreeClassifier of tree class to use Decision Tree Algorithm
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 10)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm,  annot = True)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
#Using KNeighborsClassifier Method of neighbors class to use Nearest Neighbor algorithm
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
